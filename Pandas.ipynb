Working with HUGEEE datasets

```python
new_df = pd.DataFrame(columns = df.columns)
for df in pd.read_csv('modified_pokedata.csv',chunksize = 5):
  result = df.groupby(['Type 1']).count()

  new_df = pd.concat([new_df,result])
  print(new_df)

#this means each data frame has 5 rows. i.e chunksize = 5
#breaks down large data into smaller data sets
```
